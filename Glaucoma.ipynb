```python
from sklearn.model_selection import train_test_split #training and testing data split
from sklearn import metrics #accuracy measure
from sklearn.metrics import confusion_matrix#for confusion matrix
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression,LinearRegression #logistic regression
```

```python
import pandas as pd
filename_mapping = {filename: index for index, filename in enumerate(newdata['Filename'])}
newdata['Filename'] = newdata['Filename'].map(filename_mapping)
print(newdata)
```

```python
from sklearn.preprocessing import LabelEncoder # converts gender into numbers
label_encoder = LabelEncoder()
categorical_cols_to_encode = ['Filename']
for col in categorical_cols_to_encode:
    newdata[col] = label_encoder.fit_transform(newdata[col])
newdata.head()
```

```python
train,test=train_test_split(newdata,test_size=0.3,random_state=0,stratify=newdata['Glaucoma'])
train_X=train[train.columns[:-1]]
train_Y=train[train.columns[-1:]]
test_X=test[test.columns[:-1]]
test_Y=test[test.columns[-1:]]
X=newdata[newdata.columns[:-1]]
Y=newdata['Glaucoma']
len(train_X), len(train_Y), len(test_X), len(test_Y)
```

```python
model = LogisticRegression()
model.fit(train_X,train_Y)
prediction3=model.predict(test_X)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction3,test_Y))
report = classification_report(test_Y, prediction3)
print("Classification Report:\n", report)
```

```python
model = LinearRegression()
model.fit(train_X, train_Y)
prediction = model.predict(test_X)

# Assuming 'test_Y' contains the true labels for the test set
# Calculate the accuracy
accuracy = accuracy_score(test_Y, prediction.round())

# Print the accuracy
print('The accuracy of Linear Regression is:', accuracy)
```
